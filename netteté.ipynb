```python
filtre_1 = np.array([[0, -1, 0],
                  [-1, 5, -1],
                  [0, -1, 0]])

filtre_2 = np.array([[-1, -1, -1],
                  [-1, 9, -1],
                  [-1, -1, -1]])

filtre_3 = (-1/256) * np.array([[1, 4, 6, 4, 1],
                                [4, 16, 24, 16, 4],
                                [6, 24, -476, 24, 6],
                                [4, 16, 24, 16, 4],
                                [1, 4, 6, 4, 1]])
```

```python
img_full = cv2.imread('BAS-EN-001-arkus-imperial-dragon.jpg') # shape (626, 871, 3)

sharpened_images = []
for i, filtre in enumerate([filtre_1, filtre_2, filtre_3]):
  sharpened_image = cv2.filter2D(img_full, ddepth= -1, kernel = filtre)
  # sharpened_image = cv2.filter2D(sharpened_image, ddepth= -1, kernel = filtre)
  sharpened_images.append(sharpened_image)
  cv2.imwrite(f'img_{i}.jpg', sharpened_image)
```

```python
plt.figure(figsize=(12,10))

plt.subplot(2,2,1)
plt.imshow(img_full[:,:,::-1])
plt.title('Original')

plt.subplot(2,2,2)
plt.imshow(sharpened_images[0][:,:,::-1])
plt.title('sharpened 1')

plt.subplot(2,2,3)
plt.imshow(sharpened_images[1][:,:,::-1])
plt.title('sharpened 2')

plt.subplot(2,2,4)
plt.imshow(sharpened_images[2][:,:,::-1])
plt.title('sharpened 3')

plt.show()
```
